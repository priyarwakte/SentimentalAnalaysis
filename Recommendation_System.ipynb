```python
import pandas as pd
import numpy as np
dataset = pd.read_csv('imdb_tr.csv')
dataset['user_id'] = np.random.randint(1001, 3500, dataset.shape[0])
dataset['item_id'] = np.random.randint(100,600 , dataset.shape[0])
dataset['rating'] = np.random.randint(1, 5, dataset.shape[0])

dataset.head()
```

```python
train, test = train_test_split(dataset, test_size=0.2, random_state=42)
```

```python
import torch 
import torch.nn as nn
```

```python
# creating book embedding path
movie_input = Input(shape=[1], name="Movie-Input")
movie_embedding = Embedding(1, 5, name="Movie-Embedding")(movie_input)
movie_vec = Flatten(name="Flatten-Books")(movie_embedding)
# creating user embedding path
user_input = Input(shape=[1], name="User-Input")
user_embedding = Embedding(1, 5, name="User-Embedding")(user_input)
user_vec = Flatten(name="Flatten-Users")(user_embedding)
# concatenate features
conc = Concatenate()([movie_vec, user_vec])
# add fully-connected-layers
fc1 = Dense(128, activation='sigmoid')(conc)
fc2 = Dense(32, activation='sigmoid')(fc1)
out = Dense(1)(fc2)
# Create model and compile it
model2 = Model([user_input, movie_input], out)
model2.compile('adam',loss='sparse_categorical_crossentropy')
```
